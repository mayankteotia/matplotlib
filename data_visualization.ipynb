```python
import matplotlib
import matplotlib.pyplot as plt

# Sample data (you can modify these arrays with your own data)
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [2, 4, 1, 8, 7, 12, 10, 5, 14]

# Create the scatter plot
plt.scatter(x, y, color='blue', marker='o')

# Add labels and a title
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Scatter Plot of X vs Y')

# Optional: Add gridlines
plt.grid(True)

# Show the plot
plt.show()
```

```python
#Q2 ->2. Generate a line plot to visualize the trend of values for the given data.
import matplotlib
import matplotlib.pyplot as plt

# Sample data (you can replace these with your actual data)
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [2, 4, 1, 8, 7, 12, 10, 5, 14]

# Create the line plot
plt.plot(x, y, marker='o', linestyle='-', color='green', linewidth=2)

# Add labels and title
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Line Plot of X vs Y')

# Optional: Add gridlines
plt.grid(True)

# Show the plot
plt.show()
```

```python
#Q3-> 3. Display a bar chart to represent the frequency of each item in the given array categories.
import matplotlib.pyplot as plt
from collections import Counter

# Sample data (you can replace this with your actual data)
categories = ['apple', 'banana', 'orange', 'apple', 'orange', 'apple', 'banana', 'grape']

# Count the frequency of each category
category_counts = Counter(categories)

# Extract categories and their counts for plotting
labels = list(category_counts.keys())
values = list(category_counts.values())

# Create the bar chart
plt.bar(labels, values, color='skyblue')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Frequency')
plt.title('Bar Chart of Category Frequencies')

# Optional: Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
#Q4->4. Create a histogram to visualize the distribution of values in the array data.
import matplotlib.pyplot as plt

# Sample data (you can replace this with your actual data)
data = [5, 7, 8, 5, 6, 9, 8, 7, 5, 6, 10, 9, 5, 8, 7, 10, 6, 5, 9]

# Create the histogram
plt.hist(data, bins=5, color='purple', edgecolor='black')

# Add labels and title
plt.xlabel('Value Range')
plt.ylabel('Frequency')
plt.title('Histogram of Data Distribution')

# Show the plot
plt.show()
```

```python
#Q5->5. Show a pie chart to represent the percentage distribution of different sections in the array `sections`.
import matplotlib.pyplot as plt

# Sample data (you can replace with your actual sections)
sections = ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B', 'A', 'C', 'A']
  
# Count the frequency of each section
from collections import Counter
section_counts = Counter(sections)

# Extract section labels and their frequencies
labels = list(section_counts.keys())
sizes = list(section_counts.values())

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'skyblue', 'lightgreen'])

# Add a title
plt.title('Percentage Distribution of Sections')

# Show the plot
plt.show()
```

```python
#Q6->  Create a scatter plot to visualize the relationship between two variables, by generating a synthetic
import matplotlib.pyplot as plt
import numpy as np

# Generate synthetic data
np.random.seed(42)  # For reproducibility
x = np.random.rand(50) * 100  # 50 random values between 0 and 100
y = 3 * x + np.random.randn(50) * 30  # Linear relationship with noise

# Create the scatter plot
plt.scatter(x, y, color='teal', marker='o')

# Add labels and title
plt.xlabel('X variable')
plt.ylabel('Y variable')
plt.title('Scatter Plot of Synthetic Data')

# Optional: Add a grid for better visualization
plt.grid(True)

# Show the plot
plt.show()
```

```python
#Q7->2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.
import matplotlib.pyplot as plt
import numpy as np

# Generate a dataset of 1000 random numbers from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=15, size=1000)  # Mean=50, Std Dev=15

# Create the histogram
plt.hist(data, bins=30, color='royalblue', edgecolor='black', alpha=0.7)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Randomly Generated Data')

# Show the plot
plt.show()
```

```python
#Q8->Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values
import matplotlib.pyplot as plt

# Create a dataset of categories and their corresponding values
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 12, 39, 30]

# Create the bar chart
plt.bar(categories, values, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Comparison of Categories Based on Values')

# Optional: Add gridlines for better visualization
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()

```

```python
#Q9-> Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories
import matplotlib.pyplot as plt
import numpy as np

# Generate synthetic data
np.random.seed(42)  # For reproducibility

categories = ['A', 'B', 'C', 'D']
data = {
    'A': np.random.normal(loc=50, scale=10, size=100),
    'B': np.random.normal(loc=60, scale=15, size=100),
    'C': np.random.normal(loc=55, scale=20, size=100),
    'D': np.random.normal(loc=45, scale=5, size=100)
}

# Prepare data for the box plot
values = [data[cat] for cat in categories]

# Create the box plot
plt.boxplot(values, labels=categories, patch_artist=True,
            boxprops=dict(facecolor='skyblue', color='black'))

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Distribution of Values Across Categories')

# Show the plot
plt.show()

```
